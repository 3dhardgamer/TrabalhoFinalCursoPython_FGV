```python
import pickle
import pandas as pd
import numpy as np
import datetime
import time
import sqlite3
import sqlalchemy
from IPython.display import clear_output
```

```python
format_date = lambda x: datetime.datetime.fromtimestamp(x) 
```

## Importing Data From Pickle File

```python
loaded_ohlcv_eth_usd = pickle.load(open('ohlcv_eth_usd.pkl', 'rb'))
loaded_ohlcv_eth_eur = pickle.load(open('ohlcv_eth_eur.pkl', 'rb'))
```

```python
eth_usd = pd.read_json(loaded_ohlcv_eth_usd)
eth_usd.columns = ['date', 'open', 'high', 'low', 'close', 'volume']

eth_eur = pd.read_json(loaded_ohlcv_eth_eur)
eth_eur.columns = ['date', 'open', 'high', 'low', 'close', 'volume']
```

```python
eth_usd['date'] = pd.to_datetime(eth_usd['date'].apply(format_date))
eth_eur['date'] = pd.to_datetime(eth_eur['date'].apply(format_date))
```

```python
eth_usd.set_index('date', inplace = True)
eth_eur.set_index('date', inplace = True)
```
